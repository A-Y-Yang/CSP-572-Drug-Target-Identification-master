```python
df = pd.read_csv('smiles_converted.csv')
df = df.rename(columns={'Name':'Molecule ChEMBL ID'})
```

```python
df2 = pd.read_csv('Training-Data.csv')
```

```python
new = pd.merge(df2, df, on='Molecule ChEMBL ID', how='outer')
```

```python
new = new.drop(columns=['Smiles'])
```

```python
new.head()
```

```python
y = new[{'CHEMBL614725','CHEMBL1957','CHEMBL2842','CHEMBL203'}]
```

```python
x = new.drop(columns=['CHEMBL614725','CHEMBL1957','CHEMBL2842','CHEMBL203','Molecule ChEMBL ID'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
predictions
```

```python
print("Accuracy = ",accuracy_score(y_test,predictions))
```
